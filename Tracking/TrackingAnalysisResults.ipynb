```python
import pandas as pd
import numpy as np
import plotly.express as px
import ast
import matplotlib.pyplot as plt
```

# CV Method Plots

```python
with open('Tracking_Results/trajectories_RawImage_MIL.csv') as f:
    data = f.read()
data = "{" + data + "}"
data_dict = ast.literal_eval(data)

trajectories_length = []
for vortex_no, data in data_dict.items():
    if len(data) > 1:
        trajectories_length.append(len(data))
df_lengths = pd.DataFrame(data=trajectories_length, columns=['length'])

fig = df_lengths['length'].value_counts()[:30].plot.bar(figsize=(20,10), fontsize=30)
plt.xlabel('Number of points in trajectory', fontsize=35, labelpad=15)
plt.ylabel('Value', fontsize=35, labelpad=10)
plt.grid()
```
